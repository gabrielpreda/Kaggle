# Introduction

We show Kaggle users number evolution:
* The number of new users each year  
* the number of total users (current number)  

This Notebook will be updated weekly.


```python
import pandas as pd
from datetime import datetime as dt
from IPython.display import Markdown
from IPython.display import Image
import plotly.graph_objs as go
import plotly.figure_factory as ff
from plotly import tools
from plotly.offline import download_plotlyjs, init_notebook_mode, plot, iplot
init_notebook_mode(connected=True)
n_users = pd.read_csv("../input/meta-kaggle/Users.csv")
n_users["year"] = n_users["RegisterDate"].apply(lambda x: int(x.split("/")[-1])) 
year_users = n_users.groupby("year")["Id"].count()
year_users = pd.DataFrame(year_users).reset_index()
year_users.columns = ["year", "users"]
year_users['total_users'] = year_users["users"].cumsum()
traceU = go.Bar(
    x = year_users['year'],y = year_users['users'],
    name="New users",
    marker=dict(color="Red"),
    text = year_users['users'],
)
traceT = go.Bar(
    x = year_users['year'],y = year_users['total_users'],
    name="Total users",
    marker=dict(color="Blue"),
    text = year_users['total_users'],
)

data = [traceU, traceT]
layout = dict(title = 'Kaggle Users - new users per year and total users',
          xaxis = dict(title = 'Year', showticklabels=True, dtick=1),
          yaxis = dict(title = 'Number of Kaggle users'),
          hovermode = 'closest',
          margin={"t":50,"b":0},
          height=500,
         )

fig = dict(data=data, layout=layout)
iplot(fig, filename='kaggle-users')
```

# Observations

New users dynamic on Kaggle seems to continue. This year the number of new users would probably pass 5 million, with most probably more than 20 million total users.
